 Foi escolhida uma base que contem dados sobre detecção de fraudes em pagamentos on-line.

Foi usados os dados:


*   Valor das transações
*   Tempo Entre as Transações
*   Transações nas Ultimas 24h

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("ranjitmandal/fraud-detection-dataset-csv")

print("Path to dataset files:", path)

##############################################################################



import numpy as np




print('\n\n---DADOS DE DETECÇÃO DE FRAUDES---')


#MEDIA DE VALORES DAS TRANSACOES: Ajuda entender o comportamento tipico dos clientes. Caso tenha valores muito acima ou abaixo da média podem indicar atividades suspeitas

valor_transacao = np.genfromtxt('Fraud Detection Dataset.csv', delimiter=',', usecols=[2], skip_header=1, dtype=float, encoding="ISO-8859-1", filling_values=np.nan)



valor_transacao = valor_transacao[~np.isnan(valor_transacao)]


media_vt = np.mean(valor_transacao)
print(f'\nMedia dos valores de transação:{media_vt:.2f}')


#MEDIANA DAS TRANSACOES DAS ULTIMAS 24H: Volume anormalmente alto de transações em um curto período pode ser um comportamento suspeito de fraude

transacoes_24h = np.genfromtxt('Fraud Detection Dataset.csv', delimiter=',', usecols=[9], skip_header=1, dtype=float, encoding="ISO-8859-1", filling_values=np.nan)

transacoes_24h = transacoes_24h[~np.isnan(transacoes_24h)]


mediana_t = np.median(transacoes_24h)
print(f'\nMediana das Transações nas ultimas 24h:{mediana_tt:.2f}')




#DESVIO PADRÃO DO TEMPO ENTRE AS TRANSACOES: Mede a variabilidade dos tempos. Alta variação pode indicar atividades suspeitas.


tempo_transacoes = np.genfromtxt('Fraud Detection Dataset.csv', delimiter=',', usecols=[4], skip_header=1, dtype=float, encoding="ISO-8859-1", filling_values=np.nan)

tempo_transacoes = tempo_transacoes[~np.isnan(tempo_transacoes)]

desviop_tt = np.median(tempo_transacoes)
print(f'\nDesvio Padrão do Tempo Entre as Transações:{desviop_tt:.2f}')




#MIN E MAX NUMERO DE FRAUDES: destaca os indivíduos que nunca sofreram fraudes e aqueles que enfrentaram o maior número de ocorrências, facilitando a análise de riscos


fraudes = np.genfromtxt('Fraud Detection Dataset.csv', delimiter=',', usecols=[7], skip_header=1, dtype=float, encoding="ISO-8859-1", filling_values=np.nan)

fraudes = fraudes[~np.isnan(fraudes)]

minimo = np.min(fraudes)
maximo = np.max(fraudes)


print(f'\nValor Minimo de Fraudes:{minimo:.2f}')
print(f'\nValor Maximo de Fraudes:{maximo:.2f}')










```
